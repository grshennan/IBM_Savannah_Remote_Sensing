### Using Lidar to Determine Savannah Vegetation Characteristics

This is to report on the lidar collections at Yawuru 2024

10 locations were sampled with lidar. Flight parameters were:
- 50m altitude
- 6m/s flight speed
- 50% overlap
- 5 returns

The 50% overlap parameter caused issues with later lidar thinning. This is discussed at length in the file XXXXX.

The results of this are that the lidar overlap parameter should be set to either ~20% (if canopy height only and not vegetation structure is not to be assessed), or ~70% (if vegetation structure is to be assessed).

```python
import leafmap
area_map = leafmap.Map(center=(-17.9265, 122.8843), zoom=10, height="600px")
markers = r"Map data\Study locations points.gpkg"
yawuru_area = r"Map data\Yawuru Area.gpkg"
area_map.add_vector(markers, layer_name="Study Locations")
area_style = {
    "stroke": True,
    "color": "#0000ff",
    "weight": 2,
    "opacity": 1,
    "fill": False,
}
area_map.add_vector(yawuru_area, layer_name="Yawuru Area", style=area_style)
area_map.add_layer_control()
area_map
```

From the collected lidar data several parameters are determined. The first is canopy height. The map below shows the Quarry Road scan with canopy height at the 95th percentile.
Using the 95th percentile instead of the absolute maximum excludes any noise and gives a more consistent measurement of the tops of vegetation.

```python
import leafmap.maplibregl as leafmapR
chm_map = leafmap.Map(center=(-17.724725470061554, 122.76658191658674), zoom=20, height="600px")
chm_map.add_raster(r"Map data\Yawuru_3_canopy_max.tif")
chm_map
```
